Sample Statistic: Since we usually can't measure the entire population, we collect a sample. The sample mean  is the average value from our sample. This is our best estimate of the population mean but it’s not the same thing.

Null Hypothesis (H0:μ0= μ): In hypothesis testing,we start with an assumption about the population mean we start with an assumption about the population mean ( μ0).The test’s outcome tells us whether we have enough evidence to reject this assumption based on our sample data.

Imagine you have a magic coin, and you want to know if it's a regular coin or if it's a trick coin that always lands on heads. Normally, a coin should land on *heads* or *tails* half the time, right?

So, you decide to flip the coin 10 times. And guess what? It lands on heads **every single time**!

Now, you start to think, "Hmmm... if this was just a normal coin, it should have landed on tails sometimes too." That’s where the p-value comes in. It’s like asking, “What are the chances I’d see 10 heads in a row if this coin was normal?”

If the chances are *very, very small*, it means it’s pretty unlikely this is a regular coin. The small chance (or small p-value) makes us think, "Maybe this *is* a trick coin!"

So, a p-value is just a way of seeing how likely or unlikely something is to happen if everything were normal. The smaller the chance, the more we think, “Hey, something unusual is going on!”

```python
import numpy as np
from scipy.stats import binomtest

# Set seed for reproducibility
np.random.seed(42)

# Parameters
n = 80      # number of students
x = 49      # number of correct answers
p_null = 0.5

# Use the new binomtest function
result = binomtest(x, n, p_null, alternative='two-sided')
p_value = result.pvalue
print(f"P-value: {p_value}")
```

```python
import matplotlib.pyplot as plt

# Simulate sampling distribution under H0
simulations = np.random.binomial(n, p_null, size=10000) / n
plt.hist(simulations, bins=30, color='skyblue', edgecolor='black')
plt.axvline(49 / 80, color='red', linestyle='dashed', linewidth=2)
plt.title('Sampling Distribution Under Null Hypothesis')
plt.xlabel('Proportion Correct')
plt.ylabel('Frequency')
plt.show()
```
